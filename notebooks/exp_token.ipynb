```python
import pandas as pd
import sentencepiece as spm

# Step 1: Read the CSV file
csv_file = '/content/Amharic News Dataset.csv'
df = pd.read_csv(csv_file)

# Step 2: Extract text data from the relevant column
text_data = df['article'].tolist()

# Step 3: Write the text data to a plain text file
with open('/content/data.text', 'w', encoding='utf-8') as f:
    for line in text_data:
        f.write(f"{line}\n")
```
